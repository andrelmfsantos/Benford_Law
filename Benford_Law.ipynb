```python
import pandas as pd
from benfordslaw import benfordslaw
```

```python
# Dataset:
url = 'https://raw.githubusercontent.com/andrelmfsantos/Benford_Law/main/LandingPages.csv'
df = pd.read_csv(url)

df.head(5)
```

```python
# Benfords:
#from benfordslaw import benfordslaw
# Initialize
bl = benfordslaw(alpha=0.05)

#X = df['sections'].values
#X = df['sections'].loc[(df['page'] == 'page3')].values
X = df['sections'].loc[(df['month'] == 9) & (df['page'] == 'page3')].values
print(X)

# Make fit
results = bl.fit(X)

# Plot
bl.plot(title="Benford's Law")
```
